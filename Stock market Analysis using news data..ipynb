```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#merge the datasets on the date field
merge=df1.merge(df2,how='inner',on='Date')
```

```python
#show the merged dataset
merge.head()
```

```python
#combine the top news headlines
headlines=[]
for row in range(0,len(merge.index)):
    headlines.append(''.join(str(x) for x in merge.iloc[row,2:27]))
```

```python
#print a sample of the combined headlines
headlines[0]
```

```python
#cleaning the data set.
clean_headlines=[]
for i in range(0,len(headlines)):
    clean_headlines.append(re.sub("b[(')]",'',headlines[i]))
    clean_headlines[i]=re.sub('b["]','',clean_headlines[i])
    clean_headlines[i]=re.sub("\'",'',clean_headlines[i])
```

```python
clean_headlines[0]
```

```python
#Add the clean headlines to the merge data set
merge['Combined_News']=clean_headlines
```

```python
merge['Combined_News'][0]
```

```python
merge.head(3)
```

```python
#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
#Create two new columns 'Subjectivtiy' and 'Polarity'
merge['Subjectivtiy']=merge['Combined_News'].apply(getSubjectivity)
merge['Polarity']=merge['Combined_News'].apply(getPolarity)
```

```python
#Show the new columns in the merge data set

#subjectivtiy lies between 0 and 1. (0 means factual, 1 means subjective(personal opinion))

#polarity lies between -1 and 1. (-1 means negative news and 1 means positive news)
merge.head()
```

```python
#Create a function to get the sentiment scores
def getSIA(text):
    sia = SentimentIntensityAnalyzer()
    sentiment=sia.polarity_scores(text)
    return sentiment
```

```python
#Get the sentiment scores for each day

#a score that calculates the sum of lexacon ratings that were normalized between -1 and 1
compound = []

#total is 1
neg=[]
pos=[]
neu=[]
SIA = 0

for i in range(0,len(merge['Combined_News'])):
    SIA=getSIA(merge['Combined_News'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    neu.append(SIA['neu'])
    pos.append(SIA['pos'])
```

```python
merge['Compound']=compound
merge['Negative']=neg
merge['Neutral']=neu
merge['Positive']=pos
```

```python
merge.head()
```

```python
#Create a list of columns to keep
keep_columns=['Open','High','Low','Volume','Subjectivtiy','Polarity','Compound','Positive','Negative','Label']
```

```python
df = merge[keep_columns]
```

```python
df.size
```

```python
df.head()
```

```python
#Create the feature data set
X=df
X=np.array(X.drop(['Label'],1))
y=np.array(df['Label'])
```

```python
#Split the data into 80% training and 20% testing data sets
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#train the model
model=LinearDiscriminantAnalysis().fit(x_train,y_train)
```

```python
#show model predictions (features)
predictions=model.predict(x_test)
```

```python
predictions
```

```python
#compare y_test(label) to predictions
y_test
```

```python
#Show the model metrics
print(classification_report(y_test,predictions))
```
